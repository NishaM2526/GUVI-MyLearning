```python
stack = []
stack.append(10)
stack.append(20)
stack.append(30)
stack.append(40)

top = stack.pop()
print(f"popped element :: {top}")
print(f"stack :: {stack}")
last_element = stack[-1]
print(f"last element :: {last_element}")
print(f"Is stack empty :: {len(stack)==0}")
```

```python
Queue = []
Queue.append(10)
Queue.append(20)
Queue.append(30)
Queue.append(40)
Queue.append(50)

remove = Queue.pop(0)
print(f"Removed element :: {remove}")
print(f"Queue :: {Queue}")
```

```python
#Find sum of natural numbers
num = input("Enter the number ::")
sum = 0
index = 1
while index <= int(num) :
    sum = sum + index    
    index = index + 1

print(sum)

```

```python
# Linear search

data = [10,43,56,77,83,23,65,89,77,49,99]
search = 77
index = -1
flag = False

for value in data:
    index = index + 1
    if value == search:
        print(f"Element at index {index}")
        flag = True

if flag == False:
    print("Element does not exist")
```
