```python
import numpy as np

m = np.array([
    [2,4,5],
    [4,7,9],
    [2,6,8]
])

v = np.array([2,3,6])

print("Matrix:")
print(m)
print(f"Vector : {v}")

print("Matrix multiplication :")
print(m@v)

print("Transpose of matrix :" )
print(m.T)
```

```python
print("Determinant of matrix :")
det = np.linalg.det(m)
print(det)

if det!=0:
    print("Inverse of the matrix :")
    inv = np.linalg.inv(m)
    print(inv)
else:
    print("Determinant is zero.. so inverse cannot be performed")
```

```python
import numpy as np

m = np.array([
    [2,4,5],
    [4,7,9],
    [2,6,8]
])

print("Determinant of matrix :")
det = np.linalg.det(m)
print(det)

if det!=0:
    print("Inverse of the matrix :")
    inv = np.linalg.inv(m)
    print(inv)
else:
    print("Determinant is zero.. so inverse cannot be performed")


print("Compute matrix & inverse matrix :")
m@inv
```

```python
m = np.array([
    [2,4,5],
    [4,7,9],
    [2,6,8]
])

inv = np.array([
    [-1,1,-0.5],
    [7,-3,-1],
    [-5,2,1]
])

print("Compute matrix & inverse matrix :")
m@inv
```
